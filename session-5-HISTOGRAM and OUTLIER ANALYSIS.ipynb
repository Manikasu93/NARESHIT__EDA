## OUTLIER ANALYSIS

**steps in outlieranalysis**
- step1: find the q1,q2 and q3
     - np.percentile(column data,q)
- step2: calculate the IQR
     - IQR= Q3-Q1
- step3: calculate the lower and upper boundary
     - lb: Q1-1.5IQR
     - ub: Q3-1.5IQR
- step4: Find the Outliersdf
     - c1: column data<Lb
     - c2: column data>UB
     - c: apply the main condition
     - main data[c

```python
q1=round(np.percentile(visa_df['prevailing_wage'],25),2)
q2=round(np.percentile(visa_df['prevailing_wage'],50),2)
q3=round(np.percentile(visa_df['prevailing_wage'],75),2)
IQR=(q3-q1)
ub=q3+1.5*IQR
lb=q1-1.5*IQR
con1=visa_df['prevailing_wage']>ub
con2=visa_df['prevailing_wage']<lb
con=visa_df[con1|con2]
outliers_df=con
len(outliers_df)
```

## How to deal the outliers

**Drop the outliers**
  - we can drop the outliers if outliers percentage < 25%
  - but this is not recommended , we lost other columns data also
  
**Impute with median values**
  - As we know that median does not effect by outliers
  - so it is good practice we can impute the outliers with median value
  
**cap with q3 or q1 value**
  - if outliers are persent less than lower bound then fill with Q1
  - if outliers are more than upper bound then fill with Q3

```python
#Task-3
# read the each observation from prevailing wage
# if that observation <lb or<ub :fill with median value
# else : keep as it is 


# take empty list =[]
# median= visa_df['pwage'].medain()
# for i in visa_df['pwage']:
# if i<lb or i>ub:
# emptylist.append(median)
# else:
# emptylist.append(i)
```

```python
l=[]
median=visa_df['prevailing_wage'].median()
for i in visa_df['prevailing_wage']:
    if i<lb or i>ub:
        l.append(median)
    else:
        l.append(i)
len(l)
```
