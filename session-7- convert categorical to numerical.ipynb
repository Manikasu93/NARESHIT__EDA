**One hot encoder**

- one hot encoder name saves at a time one will on and other will be off
- for example: case_sattus has two labels 
    - certified 
    - denied
- when you apply one hot encoding on case status , it creates two more extra columns
    - case_status_certified
    - case_status_denied
- one hot encoder means
    - columns are independent each other 
    - 90 degrees
    - perpendicular 90 degrees
    - orthogonal

|case_status|case_status_certified|case_status_denied|
|-|-|-|
|certified|1|0|
|denied|0|1|

**Advantage**
- when u develop ml model it is very  important the columns should be independent each other
- so here case_status creating two extra columns
- which  are independent each other, which means the row values at a time only one column has 1 

**Disadvantages**

- The disadvantages is if a columns has 100 unique labels, 100 new columns  will be created
- the data will become spares, which means huge 
- columns are more means, dimenstions are more
- the processing  times is more
- the memory consumption is more 
- *course of dimenstion*
