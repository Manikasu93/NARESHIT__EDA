```python
# conver above one into dictionary
#keys and values
#if for list
d1=dict(v_df.dtypes)
for i in d1:
    if d1[i]=='object':
        print(i)
  
```

```python
cat=[i for i in d1 if d1[i]=='object']
num=[i for i in d1 if d1[i]!='object']

print(cat)
print(num)
```

```python
cat_list=[]
num_list=[]
for i in d1:
    if d1[i]== 'object':
        cat_list.append(i)
    elif d1[i]=='int64' or d1[i]=='float64':
        num_list.append(i)
print(cat_list)
print(num_list)
```
