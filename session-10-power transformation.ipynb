```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=np.random.exponential(size=10000)
```

```python
plt.hist(data,bins=40,label='Exponential')
plt.legend()
plt.show()
# we are taking 1000 random values from expontional distrubution
# negative expontional
```

**LOG TRANSFORMATION:**

- np.log represent natural logaritham
- natural logarithm means base e
- loge(e^x)=1
- Natural logarithm will work on positive numbers only
- log transformation will not convert the data completely into normal distribution
- it will remove the skew

```python
log_data=np.log(data)
plt.hist(log_data,bins=40,label='Log Transformation')
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1).hist(data,bins=40)
plt.subplot(1,2,2).hist(data,bins=40)
plt.show()
```

```python
log_log_data=np.log(log_data)
plt.hist(log_log_data,bins=40,label='Log transformation')
plt.legend()
plt.show()
```

**RECIPROCAL TRANSFORMATION**

```python
rec_data=1/data
plt.hist(rec_data,bins=40,label='Reciprocal transformation')
plt.legend()
plt.show()
# if the data contrains 0 values , then this transformation is not useful
```

**SQUARE ROOT TRANSFORMATION**

```python
sqrt_data=data**0.5
plt.hist(sqrt_data,bins=40,label='sqrt transformation')
plt.legend()
plt.show()
```

```python
sqrt_data=np.sqrt(data)
plt.hist(sqrt_data,bins=40,label='sqrt transformation')
plt.legend()
plt.show()
```

**POWER TRANSFORMATION**

- it is in sklearn.preprocessing
- method arguments:
- box-cox
- yeo-johnson

```python
df=pd.DataFrame(data,columns=['mani'])
df
```

**BOX-COX**

```python
from sklearn.preprocessing import power_transform
pt=power_transform(df,method='box-cox')
pt
```

```python
plt.hist(pt,bins=40,label='power transformation')
plt.legend()
plt.show()
```

**yeo-johnson**

```python
from sklearn.preprocessing import power_transform
yj=power_transform(df,method='yeo-johnson')
yj
```

```python
plt.hist(yj,bins=40,label='Power transformation')
plt.legend()
plt.show()
```
